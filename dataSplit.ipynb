```python
import os
import shutil
import random

# Set paths - corrected to point to the actual data directory
base_dir = 'CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone'
output_dir = 'dataset'
split_ratio = 0.8  # 80% training, 20% testing

# Create output folders
for split in ['train', 'test']:
    for category in os.listdir(base_dir):
        # Only process directories (categories), not files
        category_path = os.path.join(base_dir, category)
        if os.path.isdir(category_path):
            os.makedirs(os.path.join(output_dir, split, category), exist_ok=True)

# Split the data
for category in os.listdir(base_dir):
    category_path = os.path.join(base_dir, category)
    
    # Only process directories (categories), not files
    if not os.path.isdir(category_path):
        continue
        
    # Get only image files, not subdirectories
    all_items = os.listdir(category_path)
    images = [item for item in all_items if os.path.isfile(os.path.join(category_path, item))]
    
    if not images:
        print(f"No images found in {category}")
        continue
        
    random.shuffle(images)

    split_point = int(len(images) * split_ratio)
    train_images = images[:split_point]
    test_images = images[split_point:]

    print(f"Processing {category}: {len(images)} images -> {len(train_images)} train, {len(test_images)} test")

    for img in train_images:
        src = os.path.join(category_path, img)
        dst = os.path.join(output_dir, 'train', category, img)
        shutil.copy(src, dst)

    for img in test_images:
        src = os.path.join(category_path, img)
        dst = os.path.join(output_dir, 'test', category, img)
        shutil.copy(src, dst)

print("Dataset successfully split into training and testing folders.")
```
